```python
#20160386 
import numpy as np
import matplotlib.pyplot as plt

def f(x0, x1):
    return x0**2 + 2*x0*x1 + 3
def df_dx(x0, x1):
    return 2*x0 + 2*x1
def delf_dx(x0, x1):
    return 2*x0 + 0*x1

x_range = 2
dx = 0.25

x0 = np.arange(-x_range, x_range+dx, dx)
x1 = np.arange(-x_range, x_range+dx, dx)
xn = x0.shape[0]

xx0, xx1 = np.meshgrid(x0, x1) #D
ff = np.zeros((len(x0), len(x1)))
dff_dx0 = np.zeros((len(x0), len(x1)))
dff_dx1 = np.zeros((len(x0), len(x1)))

for i0 in range(xn): #E
    for i1 in range(xn):
        ff[i1, i0] = f(x0[i0], x1[i1])
        dff_dx0[i1, i0] = df_dx(x0[i0], x1[i1])
        dff_dx1[i1, i0] = delf_dx(x0[i0], x1[i1])
        
plt.figure(figsize = (9,4))
#plt.subplots_adjust(wspace = 0.3)
plt.subplot(1,2,1)
cont = plt.contour(xx0, xx1, ff, 10, colors='k') #F, f 
cont.clabel(fmt = '%2.0f', fontsize = 8)

plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)

plt.subplot(1,2,2)
plt.quiver(xx0, xx1, dff_dx0, dff_dx1) #G
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)
plt.show()

plt.quiver(xx0, xx1, dff_dx0, dff_dx1) #G
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)
plt.show()
```
