```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

nx = 40
x0 = np.linspace(-5.0, 5.0, nx)
x1 = np.linspace(-5.0, 5.0, nx)

xx0, xx1 = np.meshgrid(x0, x1)
y = np.zeros((len(x0), len(x1)))

def func(x0, x1):
    x2 = np.sin(np.sqrt(x0**2 + x1**2))
    return x2

for i in range(nx):
    for j in range(nx):
        y[j, i] = func(x0[i], x1[j])
        
plt.figure(figsize=(10,7))
ax = plt.subplot(1,1,1,projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
ax.view_init(30, 70)
ax.set_zticks((0,1))
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
ax.set_zlabel('$x_2$')
plt.title('$problem2$')

plt.show()
```
