```python
#20160386이소민 기출풀기
import numpy as np
L = np.array([[2,0,0],[-3,1,0],[4,-3,7]])
U = np.array([[1,3,1],[0,1,3],[0,0,1]])

A = L.dot(U)
print(A)
```

```python
invA = np.linalg.inv(A)
print(invA)
```

```python
B = np.array([[2],[2],[3]])
result = invA.dot(B)

print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x0 = np.linspace(-5.0, 5.0, 20)
x1 = np.linspace(-5.0, 5.0, 20)

xx0, xx1 = np.meshgrid(x0, x1)

y = np.zeros((len(x0), len(x1)))

def func(x0, x1):
    res = x0**2 - x1**2
    return res

for i in range(20):
    for j in range(20):
        y[j, i] = func(x0[i], x1[j])

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3,
               color='blue', edgecolor='black')
ax.view_init(30,70)


plt.show()
```
